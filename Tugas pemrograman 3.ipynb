```python
import numpy as np

# Membuat dua array NumPy
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

# Melakukan penjumlahan
hasil = array1 + array2

# Menampilkan hasil
print("Hasil penjumlahan:", hasil)
```

```python
import pandas as pd

# Membuat DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}
df = pd.DataFrame(data)

# Menampilkan DataFrame
print("DataFrame awal:")
print(df)

# Melakukan operasi aritmatika
# Penjumlahan
df['C'] = df['A'] + df['B']

# Pengurangan
df['D'] = df['B'] - df['A']

# Perkalian
df['E'] = df['A'] * df['B']

# Pembagian
df['F'] = df['B'] / df['A']

# Menampilkan hasil
print("\nDataFrame setelah operasi aritmatika:")
print(df)
```
